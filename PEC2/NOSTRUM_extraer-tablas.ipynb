```python
##para descargarse una sola pagina de resultados
df1 = pd.DataFrame()
url = 'https://www.ebi.ac.uk/Tools/hmmer/results/48683F4A-5DFA-11EE-8529-B6E9D1021FDD/score'
#url = 'https://www.ebi.ac.uk/Tools/hmmer/results/9B283926-588A-11EE-AE85-837FD2021FDD.1/score'
html = requests.get(url).content
df_list = pd.read_html(html)
df = df_list[0]
df1 = df1.append(df)
#print(df1)
#df.to_csv('my data1.csv')
```

```python
#con este codigo vemos como generar las paginas necesarias

page = [1,2]
data = []
for i in range(1,11):
    url = f'https://www.ebi.ac.uk/Tools/hmmer/results/9B283926-588A-11EE-AE85-837FD2021FDD.1/score?page={i}'
    #print(url)
```

```python
# codigo para descargar los resultados de la pagina de hmmsearch

df2 = pd.DataFrame() #partimos de un pandas df vacio

for i in range(1,537): #en este caso hay 536 paginas con 50 resultados c/una
    url = f'https://www.ebi.ac.uk/Tools/hmmer/results/48683F4A-5DFA-11EE-8529-B6E9D1021FDD/score?page={i}'
    html = requests.get(url).content
    df_list = pd.read_html(html)
    df3 = df_list[0] #nos quedamos con el contenido de toda la tabla
    df2 = df2.append(df3) #lo anexamos al df vacio
    df2 = df2.iloc[:-1 , :] #esto elimina la ultima linea de cada pagina
print(df2)

df2.to_csv('df_fold1.csv')
```

```python
##codigo para descargar los resultados obtenidos con el HMM2_1
## disponibles en https://www.ebi.ac.uk/Tools/hmmer/results/73F0CF6E-5C70-11EE-9A99-6768D3021FDD/score

df2 = pd.DataFrame() #partimos de un pandas df vacio

for i in range(1,537): #en este caso hay 536 paginas con 50 resultados c/una
    url = f'https://www.ebi.ac.uk/Tools/hmmer/results/73F0CF6E-5C70-11EE-9A99-6768D3021FDD/score?page={i}'
    html = requests.get(url).content
    df_list = pd.read_html(html)
    df3 = df_list[0] #nos quedamos con el contenido de toda la tabla
    df2 = df2.append(df3) #lo anexamos al df vacio
    df2 = df2.iloc[:-1 , :] #esto elimina la ultima linea de cada pagina
print(df2)

df2.to_csv('df_fold4_hmm2rep2.csv')
```

```python
##codigo para descargar los resultados obtenidos con el HMM en hmmsearch con cualquier unique id

df2 = pd.DataFrame() #partimos de un pandas df vacio
unique_id = "73F0CF6E-5C70-11EE-9A99-6768D3021FDD"

for i in range(1,537): #en este caso hay 536 paginas con 50 resultados c/una
    url = f'https://www.ebi.ac.uk/Tools/hmmer/results/{unique_id}/score?page={i}'
    html = requests.get(url).content
    df_list = pd.read_html(html)
    df3 = df_list[0] #nos quedamos con el contenido de toda la tabla
    df2 = df2.append(df3) #lo anexamos al df vacio
    df2 = df2.iloc[:-1 , :] #esto elimina la ultima linea de cada pagina
print(df2)

df2.to_csv('df_fold4_hmm2rep2.csv')
```
